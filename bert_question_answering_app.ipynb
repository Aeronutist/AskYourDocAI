```python
from transformers import pipeline
```

```python

# Load pre-trained BERT Question Answering model
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

# Sample context
context = """
India is a country in South Asia. It is the seventh-largest country by area and the most populous democracy in the world.
New Delhi is the capital of India. The Indian economy is one of the world's fastest-growing major economies.
"""

# Ask a question
question = "What is the capital of India?"

# Run QA
result = qa_pipeline(question=question, context=context)
print(f"Answer: {result['answer']}")
```

```python
from google.colab import files
import pdfplumber
import os

# Upload file
uploaded = files.upload()
file_name = next(iter(uploaded))

# Auto-detect extension
ext = os.path.splitext(file_name)[-1].lower()

# Extract context from either PDF or TXT
if ext == ".txt":
    with open(file_name, "r", encoding="utf-8") as f:
        context = f.read()

elif ext == ".pdf":
    pdf_context = ""
    with pdfplumber.open(file_name) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            if text:
                pdf_context += text
    context = pdf_context

else:
    raise ValueError("Unsupported file format. Please upload a .txt or .pdf file.")

# Show preview
print("✅ Document Loaded")
print(context[:1000])  # Preview first 1000 characters
```

```python
question = " what is the main difference make it differ from other models and explain in a sentence"
result = qa_pipeline(question=question, context=context)
print("Answer:", result["answer"])
```
